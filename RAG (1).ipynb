```python
vector_store = Chroma.from_documents(
    documents=all_splits,
    embedding=embeddings,
    persist_directory="/content/chroma_vectors"
)
```

```python
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
import os
```

```python
retriever = vector_store.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)
#kwarg=keyword argument
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableLambda
from langchain_core.output_parsers import StrOutputParser

# 1. Define the prompt
prompt = PromptTemplate.from_template("""
You are a helpful assistant.

Answer the question based on the context below.

Context:
{context}

Question:
{question}

Answer:
""")

# 2. Function to format inputs for prompt
def format_inputs(inputs):
    question = inputs["question"]
    docs = inputs["context"]  # from retriever
    context = "\n\n".join(doc.page_content for doc in docs)
    return {"context": context, "question": question}

# 3. Build full RAG chain
rag_chain = (
    {
        "context": RunnableLambda(lambda d: d["question"]) | retriever,
        "question": RunnableLambda(lambda d: d["question"])
    }
    | RunnableLambda(format_inputs)
    | prompt
    | llm
    | StrOutputParser()
)

# 4. Run it
response = rag_chain.invoke({
    "question": "What is the main innovation introduced in the Attention is All You Need paper?"
})
print(response)
```
